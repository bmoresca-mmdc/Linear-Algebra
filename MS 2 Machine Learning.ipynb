```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_data, x='ProductCategory', y='Amount', hue='ProductCategory', palette='plasma', legend=False)
plt.title('Transaction Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_data, x='PaymentMethod', y='Amount', hue='PaymentMethod', palette='viridis', legend=False)
plt.title('Transaction Amount by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
pivot_table = merged_data.pivot_table(values='Amount', index='ProductCategory', columns='PaymentMethod', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Average Transaction Amount by Payment Method and Product Category')
plt.xlabel('Payment Method')
plt.ylabel('Product Category')
plt.show()
```
